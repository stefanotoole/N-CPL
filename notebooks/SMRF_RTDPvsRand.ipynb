```python
import gym
import random
```

```python
RTDPGames = {}
for name in ['alien', 'amidar', 'assault', 'asterix', 'asteroids', 'atlantis',
    'bank_heist', 'battle_zone', 'beam_rider', 'berzerk', 'bowling', 'boxing', 'breakout',
    'centipede', 'chopper_command', 'crazy_climber', 'demon_attack', 'double_dunk', 'enduro', 'fishing_derby','freeway', 'frostbite', 'gopher', 'gravitar',
    'hero', 'ice_hockey', 'jamesbond', 'kangaroo', 'krull', 'kung_fu_master',
    'montezuma_revenge', 'ms_pacman', 'name_this_game', 'phoenix', 'pitfall', 'pong',
    'private_eye', 'qbert', 'riverraid', 'road_runner', 'robotank', 'seaquest', 'skiing',
    'solaris', 'space_invaders', 'star_gunner', 'tennis', 'time_pilot', 'tutankham', 'up_n_down',
    'venture', 'video_pinball', 'wizard_of_wor', 'yars_revenge', 'zaxxon']:
    game = ''.join([g.capitalize() for g in name.split('_')])
    env = gym.make("{}-v4".format(game)).env
    env.unwrapped.frameskip = 15
    rewardsRTDP = []
    for ep in range(20):
        #print(ep)
        done = False
        step = 0
        state = env.reset()
        current_state = env.unwrapped.clone_full_state()
        total_reward = 0
        while not done and step < 50:
            maxQ = None
            nextStates = []
            for action in range(env.action_space.n): # for each action
                env.unwrapped.restore_full_state(current_state)
                state, next_reward, done, _ = env.step(action)
                next_state = env.unwrapped.clone_full_state()
                Q = next_reward
                rolloutStep = 0
                rolldone = done
                while rolloutStep < 10 and not rolldone:
                    rolloutStep += 1
                    state, reward, rolldone, _ = env.step(env.action_space.sample())
                    Q += reward
                if maxQ is None or Q > maxQ:
                    maxQ = Q
                    nextStates = [(next_state, next_reward, done)]
                elif Q == maxQ:
                    nextStates.append((next_state,next_reward, done))
            current_state, next_reward, done = random.choice(nextStates)
            total_reward += next_reward
            step += 1
        rewardsRTDP.append(total_reward)
    RTDPGames[game] = rewardsRTDP
    print(game)
    print(rewardsRTDP)
```

```python
import gym
from gym.wrappers import Monitor
for name in ['Freeway']:
    game = ''.join([g.capitalize() for g in name.split('_')])
    env = gym.make("{}-v4".format(game)).env
    env.unwrapped.frameskip = 15
    env = Monitor(env, './video_Freeway_1200_rand1')
    rewardsRandom = []
    for ep in range(1):
        #print(ep)
        done = False
        step = 0
        state = env.reset()
        env.render()
        total_reward = 0
        while not done and step < 1200:
            state, reward, done, _ = env.step(env.action_space.sample())
            env.render()
            total_reward += reward
            step += 1
        rewardsRandom.append(total_reward)
        env.close()
    print(game)
    print(rewardsRandom)
```

```python
import copy
for name in ['Freeway']:
    game = ''.join([g.capitalize() for g in name.split('_')])
    env = gym.make("{}-v4".format(game)).env
    env.unwrapped.frameskip = 15
    simEnv = copy.deepcopy(env)
    simEnv.unwrapped.frameskip = 15
    env = Monitor(env, './video_montezumas_freeway_1200_rtdp12')
    rewardsRTDP = []
    for ep in range(1):
        #print(ep)
        done = False
        step = 0
        state = env.reset()
        simEnv.reset()
        env.render()
        current_state = env.unwrapped.clone_full_state()
        total_reward = 0
        while not done and step < 1200:
            maxQ = None
            nextStates = []
            for action in range(env.action_space.n): # for each action
                simEnv.unwrapped.restore_full_state(current_state)
                state, next_reward, done, _ = simEnv.step(action)
                next_state = simEnv.unwrapped.clone_full_state()
                Q = next_reward
                rolloutStep = 0
                rolldone = done
                while rolloutStep < 10 and not rolldone:
                    rolloutStep += 1
                    state, reward, rolldone, _ = simEnv.step(env.action_space.sample())
                    Q += reward
                if maxQ is None or Q > maxQ:
                    maxQ = Q
                    nextStates = [(next_state, next_reward, done, action)]
                elif Q == maxQ:
                    nextStates.append((next_state,next_reward, done, action))
            current_state_previous = copy.deepcopy(current_state)
            current_state, next_reward, done, action = random.choice(nextStates)
            
            env.unwrapped.restore_full_state(current_state_previous)
            state, next_reward, done, _ = env.step(action)
            env.render()
            total_reward += next_reward
            step += 1
        env.close()
    print(game)
    print(rewardsRTDP)
```
